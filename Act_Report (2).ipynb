## Report: Analysis and Visualization

#### Insights

*Insight 1:* There is very little correlation between rating and either retweet count or favourite count. The ratings given by Twitter users do not show how highly rated the tweet is!

This can be seen from the following scatter matrix of all three variables.



*Insight 2:* The correlation coefficient for retweet count and favourite count is about 0.912. This is very high and shows that the variables are strongly linked. This was calculated using np.corrcoef().

*Insight 3:* Tweets are more likely to be favourited than to be retweeted. This can be seen from the following kernel density function plot, which shows the estimated probability distributions of both types of count. The favourite count is more likely to be greater, although both counts are most likely to be just above 0.



Users may possibly feel less social pressure when favouriting as this is not as public as retweeting. It does not make such a statement and so feels easier. It's therefore more popular, hypothetically.

*Insight 4:* The distribution of ratings is concentrated in the region 9.0-13.0. Above 13.0 there are very few ratings, showing this may be perceived as an upper limit by most WeRateDogs users - this would make sense as the standard upper limit would be 10 (a rating of 10/10) and so ratings of 11/10, 12/10, 13/10 are deliberately outlandish (and part of the attraction of WeRateDogs).



*Insight 5:* The character count of each tweet is also skewed right, with a fair number of tweets above 140 characters. However there is virtually no correlation between character count and any of the other variables favourite count, retweet count or rating.
